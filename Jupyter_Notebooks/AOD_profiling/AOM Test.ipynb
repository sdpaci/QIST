```python
#Defines array for number of entries in each bin for baseline

#returns as a tuple
x1bY = plt.hist(x1b, bins = 256, range = (0, 256))

#converts tuple to list
x1bYList = list(x1bY[0])
x1bYI = [int(i) for i in x1bYList]

#Creates lists of background as X and Y points
X1BX = []; X1BY = []
for i in range(256):
    if x1bYI[i] != 0:
        X1BY.append(x1bYI[i])
        X1BX.append(i)
    else:
        print(i)

print (len(X1BX), len(X1BY))
```

```python
#Finds line of best fit for arrays based on entire array

bf1 = np.polyfit(X1BX, X1BY, 1)
print(bf1)

bfx1 = np.linspace(0, 256, 10000)
def f(bfx1): return bf1[0]*bfx1 + bf1[1]
```

```python
plt.rcParams["figure.figsize"] = (8,8)


plt.scatter(X1BX, X1BY, label = 'GaAs1')
plt.plot(bfx1, f(bfx1))
plt.fill_between(bfx1, f(bfx1), where = [(bfx1 >= x1min) and (bfx1 <= x1max) for bfx1 in bfx1], alpha = 0.3)
plt.fill_between(bfx1, f(bfx1), where = [(bfx1 >= x2min) and (bfx1 <= x2max) for bfx1 in bfx1], alpha = 0.3)
plt.legend()
plt.show()
```

```python
bf1deg2 = np.polyfit(X1BX, X1BY, 2)
print(bf1deg2)

bf1xdeg2 = np.linspace(0, 256, 10000)
def g(bf1xdeg2): return bf1deg2[2] + bf1deg2[1]*bf1xdeg2 + bf1deg2[0] * bf1xdeg2**2
```

```python
plt.rcParams["figure.figsize"] = (8,8)

#plt.scatter(X1BX, X1BY, label = 'GaAs1')
plt.plot(bf1xdeg2, g(bf1xdeg2), color = 'red')
#plt.fill_between(bf1xdeg2, g(bf1xdeg2), where = [(bf1xdeg2 >= x1min) and (bf1xdeg2 <= x1max) for bf1xdeg2 in bf1xdeg2], color = 'blue', alpha = 0.3)
#plt.fill_between(bf1xdeg2, g(bf1xdeg2), where = [(bf1xdeg2 >= x2min) and (bf1xdeg2 <= x2max) for bf1xdeg2 in bf1xdeg2], color = 'blue', alpha = 0.3)
plt.ylim(0,120000)
plt.show()
```

```python
import sympy as sy

bf1xdeg2 = sy.Symbol('bf1xdeg2')

area_Under_1 = sy.integrate(g(bf1xdeg2), (bf1xdeg2, x1min, x1max))
print (area_Under_1)

area_Under_2 = sy.integrate(g(bf1xdeg2), (bf1xdeg2, x2min, x2max))
print (area_Under_2)
```

```python
Peak_Area_1 = len(x1p1) - area_Under_1
print ('Area under first peak, subtracting background is', Peak_Area_1)
Peak_Area_2 = len(x1p2) - area_Under_2
print ('Area under second peak, subtracting background is', Peak_Area_2)
```
