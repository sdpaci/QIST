```python
# read a data file, store all data in arrays

file = "/Users/steve/AOM/100kHz-500ns-Pulse_Isolate8x8_10s_W0028_F03-191206-153211-1_cent.csv"
#file = "/Users/andreinomerotski/Andrei/python/WSUdata/06142018_500k_run1_W0028_H11-180614-145223-1.csv"
print (file)
print(time.ctime())

data=np.loadtxt(file, dtype= float, delimiter=",", usecols = (0,1,2,3))

y   = data[:, 0]
x   = data[:, 1]
t   = data[:, 2]
a   = data[:, 3]

print (len(t), time.ctime())
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(10, 3))

ax0.hist(a, bins = 100, range = (0, 2500), color = 'r', ec = 'k')
plt.title("TOT", fontsize = 12) # change the title
plt.xlabel('TOT(ns)',fontsize = 12)
plt.ylabel('counts',fontsize = 12)
#plt.xlim(0, 240)
#plt.yscale('log')
#plt.ylim(0, 10000)
#plt.show()

time.ctime()
```
