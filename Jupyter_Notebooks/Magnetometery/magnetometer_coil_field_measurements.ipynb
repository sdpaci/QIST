```python
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import glob
import time
import pandas as pd
from statistics import mean
import math
from scipy import constants

# plt.rcParams["figure.figsize"] = (3,3)
print (time.ctime())
```

```python
current = [0,4,10,20,30,60,80,100]
# old data
# P1 = [30,80,70,40,160,200,260,290]
# P2 = [30,70,70,50,170,220,320,380]
# P3 = [40,30,80,60,140,250,280,350]
# P4 = [90,120,170,0,250,290,350,480]
# P5 = [10,70,10,80,170,280,310,470]
# P6 = [0,60,60,80,140,250,290,400]
# new data
P1 = [60,80,100,120,140,240,280,350]
P2 = [30,70,90,100,130,250,310,380]
P3 = [60,70,60,90,120,250,300,380]
P4 = [80,100,140,160,180,300,350,350]
P5 = [40,70,90,140,150,270,320,400]
P6 = [10,40,70,90,150,230,320,360]
```

```python
plt.rcParams["figure.figsize"] = (20,6)
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'

plt.plot(current,P1, '.', linestyle='--', linewidth=1)
plt.plot(current,P2, '.', linestyle='--', linewidth=1)
plt.plot(current,P3, '.', linestyle='--', linewidth=1)
plt.plot(current,P4, '.', linestyle='--', linewidth=1)
plt.plot(current,P5, '.', linestyle='--', linewidth=1)
plt.plot(current,P6, '.', linestyle='--', linewidth=1)
plt.title("Magnetometer Coil Field at 6 Positions in Rb Cell Location", fontsize = 12) # change the title
plt.xlabel('Current [mA]',fontsize = 16)
plt.ylabel('Magnetic Field [mG]',fontsize = 16)
plt.xticks(current, fontsize = 16, rotation="vertical")
plt.yticks(fontsize=12)
plt.grid()
plt.legend(["Pos. 1", "Pos. 2", "Pos. 3","Pos.4","Pos. 5","Pos. 6"])

plt.savefig('magnetometer_coil_field_plot.png', bbox_inches='tight')
time.ctime()
```

```python
#Theoretical B = Mu*n*I

mu=constants.mu_0
mu_conversion = mu*10000 # (T*m/A) -> (G*m/A)
n=198
muN = mu_conversion*n
B = [muN*I for I in current]
print(B)
```

```python
# Gauss for optical pumping

I  = (10*(0.1297686))/(muN)
I
```

```python
L = 0.1297686
B_L = [x/L for x in B]
B_L
```

```python
plt.rcParams["figure.figsize"] = (20,6)
plt.rcParams['axes.facecolor']='white'
plt.rcParams['savefig.facecolor']='white'

plt.plot(current,P1, '.', linestyle='--', linewidth=1)
plt.plot(current,P2, '.', linestyle='--', linewidth=1)
plt.plot(current,P3, '.', linestyle='--', linewidth=1)
plt.plot(current,P4, '.', linestyle='--', linewidth=1)
plt.plot(current,P5, '.', linestyle='--', linewidth=1)
plt.plot(current,P6, '.', linestyle='--', linewidth=1)
plt.plot(current,B_L, '.', linestyle='-', linewidth=1)
plt.title("Magnetometer Coil Field at 6 Positions in Rb Cell Location", fontsize = 12) # change the title
plt.xlabel('Current [mA]',fontsize = 16)
plt.ylabel('Magnetic Field [mG]',fontsize = 16)
plt.xticks(current, fontsize = 16, rotation="vertical")
plt.yticks(fontsize=12)
plt.grid()
plt.legend(["Pos. 1", "Pos. 2", "Pos. 3","Pos.4","Pos. 5","Pos. 6", "Theoretical"])

plt.savefig('magnetometer_coil_field_plot_data_and_theoretical.png', bbox_inches='tight')
time.ctime()
```
